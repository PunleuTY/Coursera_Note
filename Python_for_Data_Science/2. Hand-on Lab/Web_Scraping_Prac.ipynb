# First Hand on Practicing Web Scraping

So, for this practice, i will focus on the Beautiful Soup Library. This library is used to extract the info from HTML, XML and other format, but in this jupyter, i will mainly dive into the HTML part, how to apply the bs4 to extract data from HTML. 

- bs4 has ability to tranform complex HTML docutment into the **complex tree python** objects 


```python
<!DOCTYPE html>
<html>
<head>
<title>Page Title</title>
</head>
<body>
<h3><b id='boldest'>Lebron James</b></h3>
<p> Salary: $ 92,000,000 </p>
<h3> Stephen Curry</h3>
<p> Salary: $85,000, 000 </p>
<h3> Kevin Durant </h3>
<p> Salary: $73,200, 000</p>
</body>
</html>
```

Imagine that our target website has the HTML structure like above, and we would like to extract the data from this website without copy & paste each info one-by-one. SO what we can do is using bs4. 

There are 2 ways to specify the HTML to extracted, 
- First we can copy the structure of HTML and declare a variable to store it 
- OR we can import the requests module to working with the URL and access the HTML structure through URL, then extracted (Recommeded)

The second option is the best but now i've just started and to make it clear for me as a starter, i'd like to go with first option. 

```python
from bs4 import BeautifulSoup


html = "<!DOCTYPE html><html><head><title>Page Title</title></head><body><h3><b id='boldest'>Lebron James</b></h3><p> Salary: $ 92,000,000 </p><h3> Stephen Curry</h3><p> Salary: $85,000, 000 </p><h3> Kevin Durant </h3><p> Salary: $73,200, 000</p></body></html>"

soup = BeautifulSoup(html, 'html.parser')
# so this above variable, as we can see the HTML is parsed and stored in the soup variable as the tree python object

# WE can also display the nice structure of HTML as well by using the prettify() method
print(soup.prettify())
```
